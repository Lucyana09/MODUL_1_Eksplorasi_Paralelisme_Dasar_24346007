```python
import multiprocessing
import random
import time
import math

def monte_carlo_chunk(chunk_size):
    inside = 0
    for _ in range(chunk_size):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        if x**2 + y**2 <= 1:
            inside += 1
    return inside

def calculate_pi_parallel(num_samples, num_processes):
    chunk_size = num_samples // num_processes
    with multiprocessing.Pool(processes=num_processes) as pool:
        results = pool.map(monte_carlo_chunk, [chunk_size] * num_processes)
    total_inside = sum(results)
    return 4 * total_inside / num_samples

if __name__ == "__main__":
    num_samples = 10_000_000
    process_counts = [1, 2, 4, 8, 16, 32]
    
    print("PROGRAM PARALEL - MULTIPROCESSING")
    print(f"Jumlah sampel: {num_samples:,}")
    print(f"{'Proses':^8} | {'Waktu (detik)':^12} | {'Estimasi π':^16}")
    print("-" * 70)
    
    for num_procs in process_counts:
        start_time = time.time()
        pi_est = calculate_pi_parallel(num_samples, num_procs)
        end_time = time.time()
        
        execution_time = end_time - start_time
        
        print(f"{num_procs:^8} | {execution_time:^12.4f} | {pi_est:^16.8f}")
    
    print("-" * 70)
```
