```python
import matplotlib.pyplot as plt

processes = [1, 2, 4, 8, 16, 32]
waktu = [3.6985, 2.3300, 2.1864, 2.2008, 2.4173, 2.3565]
serial_time = 3.6006  

# Hitung speedup dan efficiency
speedups = [serial_time / t for t in waktu]
efficiencies = [speedups[i] / processes[i] for i in range(len(processes))]
ideal = processes

plt.figure(figsize=(10, 6))
plt.plot(processes, speedups, 'bo-', linewidth=2, markersize=8, label='Speedup Aktual')
plt.plot(processes, ideal, 'r--', linewidth=2, label='Speedup Ideal')
plt.xlabel('Jumlah Proses', fontsize=12)
plt.ylabel('Speedup', fontsize=12)
plt.title('Speedup vs Jumlah Proses', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.xticks(processes)
plt.savefig('speedup_grafik.png', dpi=150)
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(processes, efficiencies, 'go-', linewidth=2, markersize=8)
plt.axhline(y=1.0, color='gray', linestyle='--', alpha=0.5)
plt.xlabel('Jumlah Proses', fontsize=12)
plt.ylabel('Efficiency', fontsize=12)
plt.title('Efficiency vs Jumlah Proses', fontsize=14)
plt.grid(True, alpha=0.3)
plt.xticks(processes)
plt.savefig('efficiency_grafik.png', dpi=150)
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(processes, waktu, 'mo-', linewidth=2, markersize=8, color='purple')
plt.axhline(y=serial_time, color='red', linestyle='--', alpha=0.7, label=f'Waktu Serial ({serial_time} detik)')
plt.xlabel('Jumlah Proses', fontsize=12)
plt.ylabel('Waktu Eksekusi (detik)', fontsize=12)
plt.title('Waktu Eksekusi vs Jumlah Proses', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.xticks(processes)
plt.savefig('waktu_grafik.png', dpi=150)
plt.show()


print("HASIL ")
print(f"{'Proses':^8} | {'Waktu (s)':^10} | {'Speedup':^10} | {'Efficiency':^10}")
print("-"*65)
for i, p in enumerate(processes):
    print(f"{p:^8} | {waktu[i]:^10.4f} | {speedups[i]:^10.4f} | {efficiencies[i]:^10.4f}")
print("="*65)
```
